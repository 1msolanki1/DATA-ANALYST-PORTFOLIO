```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Get the user's home directory
home_dir = os.path.expanduser("~")

# Construct the full path to the file in the Downloads folder
file_path = os.path.join(home_dir, "Desktop","Portfolio","Python","top-5000-youtube-channels.csv")

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
data.info()
```

```python
# Overall Statistics about the dataframe
data.describe()
```

```python
data.dtypes
```

```python
# Data Cleaning
data.head(20)
```

```python
import numpy as np
```

```python
data=data.replace('--',np.nan,regex=True) 
data.head(20)
```

```python
#checking for null values
data.isnull().sum()
```

```python
per_missing=data.isnull().sum()* 100/len(data)
per_missing
```

```python
sns.heatmap(data.isnull())
```

```python
#dropping the missing values
data.dropna(axis=0,inplace=True)
sns.heatmap(data.isnull())
```

```python
data['Rank']=data['Rank'].str[0:-2] # To remove the string 
```

```python
data['Rank']=data['Rank'].str.replace(',','').astype('int') #Replace and convert to integer 
```

```python
data['Video Uploads']=data['Video Uploads'].astype('int')  #Convert to integer 
data['Subscribers']=data['Subscribers'].astype('int')  #Convert to integer
```

```python
data.dtypes
```

```python
data['Grade'].unique()
```

```python
data['Grade']=data['Grade'].map({'A++ ':5, 'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
```

```python
data.head()
```

```python
data.dtypes
```

```python
# Calculate average views per video upload
data['Avg_views'] = data['Video views'] / data['Video Uploads']
data['Avg_views'] = data['Avg_views'].astype(int)
data.head()
```

```python
# Sorting Top 5 channels with max no of video uploads
data.sort_values(by= 'Video Uploads', ascending=False).head()
```

```python
# Calculating correlation Matrix/ Correlation Analysis
# Compute pairwise correlation
correlation_matrix = data[['Video Uploads', 'Subscribers', 'Video views']].corr(method='pearson')

# Visualize correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

Subscribers and Video views are highly correlated which means if there are high no of Subscribers than it will lead to high no of views. 

```python
# Which Grade has the highest number of Video Uploads
plt.figure(figsize=(8, 6))
sns.barplot(x='Grade', y='Video Uploads', data=data)
plt.title('Grade with Highest Number of Video Uploads')
plt.xlabel('Grade')
plt.ylabel('Video Uploads')
plt.show()

# Which Grade has the highest average views
plt.figure(figsize=(8, 6))
sns.barplot(x='Grade', y='Avg_views', data=data)
plt.title('Grade with Highest Average Views')
plt.xlabel('Grade')
plt.ylabel('Average Views')
plt.show()

# Which Grade has the highest subscribers
plt.figure(figsize=(8, 6))
sns.barplot(x='Grade', y='Subscribers', data=data)
plt.title('Grade with Highest Subscribers')
plt.xlabel('Grade')
plt.ylabel('Subscribers')
plt.show()
```


The performance metrics by grade indicate a clear trend of improvement across all variables as the grade increases. Grade 5 that is A++ exhibit the highest values in most metrics, with significantly higher numbers of video uploads, subscribers, and video views compared to lower grades. Interestingly, Grade 4 that is A+ have the highest average views per video upload, suggesting higher engagement despite fewer uploads. These findings suggest a positive correlation between grade and channel success, with higher-grade channels demonstrating superior performance across the board.



```python
#Top Channels Analysis
# Visualize distributions using histograms
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.histplot(data['Video Uploads'], kde=True, color='skyblue')
plt.xlabel('Video Uploads')
plt.title('Distribution of Video Uploads')

plt.subplot(1, 3, 2)
sns.histplot(data['Subscribers'], kde=True, color='salmon')
plt.xlabel('Subscribers')
plt.title('Distribution of Subscribers')

plt.subplot(1, 3, 3)
sns.histplot(data['Video views'], kde=True, color='green')
plt.xlabel('Video views')
plt.title('Distribution of Video views')

plt.tight_layout()
plt.show()

# Summary statistics
summary_stats = data[['Video Uploads', 'Subscribers', 'Video views']].describe()
print(summary_stats)



```

```python
# Step 1: Identify the top channels based on 'Video Uploads', 'Subscribers', and 'Video views'

# Top channels based on 'Video Uploads'
top_channels_uploads = data[['Channel name', 'Video Uploads']].sort_values(by='Video Uploads', ascending=False).head(10)

# Top channels based on 'Subscribers'
top_channels_subscribers = data[['Channel name', 'Subscribers']].sort_values(by='Subscribers', ascending=False).head(10)

# Top channels based on 'Video views'
top_channels_views = data[['Channel name', 'Video views']].sort_values(by='Video views', ascending=False).head(10)

# Step 2: Compare the performance metrics of the top channels across different grades

# Group by 'Grade' and calculate mean values for each metric
top_channels_grades = data.groupby('Grade').agg({
    'Video Uploads': 'mean',
    'Subscribers': 'mean',
    'Video views': 'mean'
}).reset_index()

# Print the top channels and their grades
print("Top Channels based on Video Uploads:")
print(top_channels_uploads)
print("\nTop Channels based on Subscribers:")
print(top_channels_subscribers)
print("\nTop Channels based on Video Views:")
print(top_channels_views)
print("\nPerformance Metrics of Top Channels across Different Grades:")
print(top_channels_grades)



```

Channels like AP Archive, PewDiePie, and T-Series excelled in these metrics, showcasing prolific content output and large subscriber bases. Higher-grade channels demonstrated superior performance across all metrics, indicating a positive correlation between grade and channel success.

```python
# Outlier Detection:

from sklearn.ensemble import IsolationForest

# Detect outliers using Isolation Forest
outlier_detector = IsolationForest(contamination=0.05)
outliers = outlier_detector.fit_predict(data[['Video Uploads', 'Subscribers', 'Video views']])

# Mark outliers in the dataframe
data['Outlier'] = outliers
print("Outlier Detection Results:")
print(data[data['Outlier'] == -1])


```

The outlier detection process identified 231 channels exhibiting anomalous behavior in terms of video uploads, subscribers, and video views compared to the rest of the dataset.
